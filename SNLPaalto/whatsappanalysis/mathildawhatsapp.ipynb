```python
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English

# Create a custom tokenizer with English vocabulary
nlp = English()
tokenizer = Tokenizer(nlp.vocab)


def get_20_most_common_words(text, pos_tags=["NOUN", "VERB", "ADJ"]):
    """
    Extracts the 20 most common words from a text string, filtering by specified part-of-speech tags.

    Args:
        text: The text string to analyze.
        pos_tags: A list of part-of-speech tags to filter by (e.g., NOUN, VERB, ADJ).

    Returns:
        A set containing the 20 most common words (excluding stop words).
    """
    # Tokenize the text
    doc = nlp(tokenizer(text))
    
    # Filter tokens by part-of-speech tags (if provided)
    #tokens = [token for token in doc if token.pos_ in pos_tags]
    

    # Exclude stop words (adjust stop word list as needed)
    stop_words = spacy.lang.en.stop_words.STOP_WORDS
    tokens = [token.text.lower() for token in doc if token.text.lower() not in stop_words]

    # Get word frequencies
    from collections import Counter
    word_counts = Counter(tokens)

    # Return the 20 most common words
    return set(word_counts.most_common(20))  # Adjust the number of most common words as desired

result = line  # Assuming "line" is your English chat text

resultado = set()
chunk_size = 100000  # Adjust chunk size as needed based on memory and processing limitations

for i in range(0, len(result), chunk_size):
    # Process text in chunks to avoid memory issues with large datasets
    if i == 200000:
        text_chunk = result[i:i+chunk_size]
        most_common_words = get_20_most_common_words(text_chunk)
        resultado.update(most_common_words)  # Update the set with new words
        

    else: 
        text_chunk = result[i:]
        most_common_words = get_20_most_common_words(text_chunk)
        resultado.update(most_common_words)

# Save the results to a text file
print(resultado)
with open("result.txt", "w", encoding="utf-8") as file:
    for elem in resultado:
        file.write(elem + "\n")
```
